```python
#uniform distribution 
#used to describe the propability where every event has equal chances of occurance 
# Example :create a 2x3 uniform distribution 
from numpy import random 
x=random.uniform(size=(2,3))
print(x)
```

```python
#visualisation of uniform distribustion 
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.uniform(size=1000), hist=False)

plt.show()
```
